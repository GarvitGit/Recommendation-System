```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("data/wikipedia/wikipedia-vectors.csv")
df
```

```python
dp=pd.read_csv("data/wikipedia/wikipedia-vocabulary-utf8.txt",names=["Words"])
dp
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df.head()
```

```python
df=df.T
```

```python
df.head()
```

```python
from sklearn.decomposition import NMF
```

```python
nmf=NMF(n_components=6)
```

```python
nmf.fit(df)
```

```python
titles=list(df.T.columns)
```

```python
x=nmf.transform(df)
```

```python
titles
```

```python
from sklearn.cluster import KMeans
```

```python
knn=KMeans(n_clusters=10)
```

```python
knn.fit(x)
```

```python
knn.predict(x)
```

```python
kl=pd.DataFrame({"Group":knn.predict(x),"prediction":titles})
```

```python
kl.sort_values(by=["Group"])
```
